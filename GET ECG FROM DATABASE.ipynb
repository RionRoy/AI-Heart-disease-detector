```python
import numpy as np
import os
import shutil
import random

# 1. Load the Data (Assuming Drive is already mounted)
print("â³ Loading dataset... this might take a moment.")
data = np.load("/content/drive/MyDrive/PTBXL/ptbxl_train_preprocessed_100hz_10s.npz", allow_pickle=True)
X_train = data['X']
y_train = data['y']

# 2. Create a temporary folder
folder_name = "100_Test_ECGs"
if os.path.exists(folder_name):
    shutil.rmtree(folder_name) # Clean up if exists
os.makedirs(folder_name)

# 3. Pick 100 Random Indices
total_patients = len(X_train)
random_indices = np.random.choice(total_patients, 100, replace=False)
print(f"âœ… Selected 100 random patients from {total_patients} total records.")

# 4. Save Files & Create an Answer Key
answer_key_path = os.path.join(folder_name, "Answer_Key.txt")

with open(answer_key_path, "w") as f:
    f.write(f"ID | TRUE DIAGNOSIS CODE\n")
    f.write("-" * 30 + "\n")

    for idx in random_indices:
        # Save the ECG file
        filename = f"patient_{idx}.npy"
        file_path = os.path.join(folder_name, filename)
        np.save(file_path, X_train[idx])

        # Write the true diagnosis to the text file
        # (This helps you verify if your App is correct)
        diagnosis = y_train[idx]
        f.write(f"{filename} : {diagnosis}\n")

print("âœ… Saved 100 .npy files.")

# 5. Zip it into a single download
print("ğŸ“¦ Zipping files...")
shutil.make_archive("/content/100_random_ecgs", 'zip', folder_name)

print("\nğŸ‰ SUCCESS!")
print("ğŸ‘‰ Go to the Files folder on the left (ğŸ“ icon).")
print("ğŸ‘‰ Right-click '100_random_ecgs.zip' and click DOWNLOAD.")
```
